```python
# 데이터 불러오기

from datasets import load_dataset

dataset = load_dataset("beomi/KoAlpaca-v1.1a")
dataset
```

```python
# 데이터셋에서 랜덤으로 n개 뽑기

n = 10
sample_dataset = dataset["train"].select(range(n))
sample_dataset
```

3. map에 적용할 함수 만들기

```python
def make_prompt(data):
    return {
        "test": [
            {"role": "user", "content": data["instruction"]},
            {"role": "assistant", "content": data["output"]},
        ]
    }
```

```python
sample_dataset = sample_dataset.map(make_prompt)
print(sample_dataset[0])
```

# daekeun-ml/naver-news/summarization-ko 데이터

```python
# 데이터 불러오기

from datasets import load_dataset

dataset = load_dataset("daekeun-ml/naver-news-summarization-ko")
dataset
```

```python
# 데이터 1개 확인하기

train_dataset = dataset["train"][0]
test_dataset = dataset["test"][0]
validation_data = dataset["validation"][0]
```

```python
# 데이터셋에서 랜덤으로 n개 뽑기

n = 10
sample_dataset1 = dataset["train"].select(range(n))
sample_dataset2 = dataset["test"].select(range(n))
sample_dataset3 = dataset["validation"].select(range(n))
print(sample_dataset1)
print(sample_dataset2)
print(sample_dataset3)
```

```python
def make_prompt(data):
    return {
        "test": [
            {"role": "user", "content": data["instruction"]},
            {"role": "assistant", "content": data["output"]},
        ]
    }
```

```python
sample_dataset = sample_dataset.map(make_prompt)
print(sample_dataset1[0])
print(sample_dataset2[0])
print(sample_dataset3[0])
```
